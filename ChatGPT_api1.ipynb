# Basic Chat with openai python library

```python
# create a basic chatbot on an infinite loop where the user can insert a question, it sends it to chatGPT using the api and it fetches the response
while True:
    # get a message from the user
    question = input("User: ")
    if question:
        # if the question is "quit", then terminate the loop
        if question.lower() == "quit":
            break
        # append the given question to the sequence of questions making up the conversation
        messages.append(
            {
                "role": "user",
                "content": question,
            }
        )
        # send the whole conversation to chatGPT (don't forget to pick the appropriate model)
        chat = openai.ChatCompletion.create(
            model="gpt-3.5-turbo", messages=messages
        )
    # the reply we'll work with will be the first message generated by chatGPT
    reply = chat.choices[0].message.content
    print(f"Bot: {reply}")
    # Don't forget to append the message to the conversation
    messages.append(
        {
            "role": "assistant",
            "content": reply,
        }
    )
```

# Basic chat with POST request for OpenAI endpoint

```python
import requests

URL = "https://api.openai.com/v1/chat/completions"

payload = {
"model": "gpt-3.5-turbo",
"messages": [{"role": "user", "content": f"What is the first computer in the world?"}],
"temperature" : 1.0,
"top_p":1.0,
"n" : 1,
"stream": False,
"presence_penalty":0,
"frequency_penalty":0,
}

headers = {
"Content-Type": "application/json",
"Authorization": f"Bearer {openai.api_key}"
}

response = requests.post(URL, headers=headers, json=payload, stream=False)
     
```

```python
response.content
```

```python
# Cast a bytes object to a JSON dictionary
import json
json_bytes = response.content
json_dict = json.loads(json_bytes)
```
